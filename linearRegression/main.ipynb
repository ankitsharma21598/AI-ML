```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
import os
os.getcwd()

```

```python
df = pd.read_csv("./statsAndProbab/pizza_sales_data.csv")

print(df.head())
```

```python
# Convert pizza_size to numeric (modify this mapping based on your data)
size_map = {"S": 8, "M": 10, "L": 12, "XL": 14, "XXL": 16}
df["pizza_size"] = df["pizza_size"].map(size_map)
print(df.head(10))

# Features and target
x=df[['pizza_size']]
y=df[['unit_price']]

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
model=LinearRegression()
model.fit(x_train,y_train)

```

```python
predictions = model.predict(x_test)

print("predictions",predictions)

print("Coefficient==>", model.coef_)
print("intercepter==>", model.intercept_)
```
